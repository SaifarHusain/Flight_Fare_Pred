2.By addressing outliers and encoding issues, we’re reducing noise and helping the model focus on meaningful patterns.

```python
plt.figure(figsize=(30,50))   # Show the Outliers
plotnum=1  
for column in data.columns[2:]:
    if plotnum<=57:
        plt.subplot(15,4,plotnum)
        sns.boxplot(x=data[column])
        plt.xlabel(column,fontsize=25)
        plt.xticks(fontsize=15)
        plt.yticks(fontsize=15)
    plotnum+=1
plt.tight_layout()
```

```python
# Use the IQR Method
for column in data.drop(['ps_ind_05_cat','ps_ind_08_bin', 'ps_ind_09_bin', 'ps_ind_10_bin', 'ps_ind_11_bin',
       'ps_ind_12_bin', 'ps_ind_13_bin', 'ps_ind_14','ps_ind_17_bin', 'ps_ind_18_bin','ps_car_02_cat','ps_car_04_cat','ps_car_07_cat','ps_car_08_cat',
                        'ps_car_10_cat','ps_calc_15_bin','ps_calc_20_bin','id','target'],axis=1):
    Q1=data[column].quantile(0.25)
    Q3=data[column].quantile(0.75)
    IQR=Q3-Q1
    lower_boundary=Q1-1.5*IQR
    upper_boundary=Q3+1.5*IQR
    data.loc[(data[column]>upper_boundary) | (data[column]<lower_boundary),column]=np.median(data[column])
```

```python
from sklearn.model_selection import train_test_split   # import train test split method
from imblearn.over_sampling import SMOTE   #import Over_Sampling method
from imblearn.under_sampling import RandomUnderSampler   #import Under_Sampling method
from collections import Counter  #import Counter for count
```

#### OverSampling

1.Applied SMOTE to balance the minority class by synthetically generating new samples.

2.To prevent the model from being biased toward the majority class.

```python
sm=SMOTE(random_state=7)   # Call the SMOTE
X_sm,y_sm=sm.fit_resample(X,y)
```

```python
print("After SMOTE Oversampling:", Counter(y_sm))
```

#### UnderSampling

1.Used RandomUnderSampler to reduce samples from the majority class.

2.To reduce training time and memory usage.

3.To try a different balancing approach and compare with SMOTE.

```python
us=RandomUnderSampler(random_state=7)  #Call the Under_Sampling
X_us,y_us=us.fit_resample(X,y)
```

```python
print("After Random Undersampling:", Counter(y_us))
```

```python
# create the training data and testing data
X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.2, random_state=7)
X_train1, X_test1, y_train1, y_test1 = train_test_split(X_us, y_us, test_size=0.2, random_state=7)

```
